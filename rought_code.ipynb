```python
def recursive_path(doc, token, path, dist, too_pos):
    # path += "(Word:" + token.text + "  POS:" + token.pos_ + " DEP: " + token.dep_ + ")"
    
    # if current token is of the POS being referenced return it 
    if (token.pos_ in too_pos):
        return dist, token.text
    else:
        #visit children or current token too look for POS tag 
        dist1, word1 = recurse_down(doc, token, path,  dist + 1, too_pos)

        # visit parent of current node 
        if token.dep_ != "ROOT":
            dist2, word2 = recursive_path(doc, token.head, path, dist + 1, too_pos)
            if dist1 == 0 or (dist1 != 0 and dist2 != 0 and dist2 < dist1):
                return dist2, word2
        
        return dist1, word1
        



def recurse_down(doc, token, path, dist, too_pos):
    dist1, word1 = 0, ""

    for child in token.children:
        if child.pos_ in too_pos:
            return (dist, child.text)
        else:
            dist2, word2 =  recurse_down(doc, child, path, dist + 1, too_pos)
            if dist1 == 0 or (dist1 != 0 and dist2 != 0 and dist2 < dist1):
                dist1, word1 = dist2, word2
    
    return dist1, word1


        
def find_relations(doc, from_pos, too_pos):   
    
    # Iterate over each token in the sentence
    relations = {}
    for token in doc:
        # If the token is a noun or named entity, add it to the dictionary
        if token.pos_ in from_pos:
            dist, ref = recursive_path(doc, token, "", 0, too_pos)
            if token.text in relations:
                relations[token.text].append(ref)
            else:
                relations[token.text] = [ref]
            print(token.text + "->" + ref + " (" + str(dist) + ")")
        
    return relations
```
